```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 #read the csv file into a dataframe'
df = pd.read_csv('JMpipeline_V7.2Image.csv')
df
```

```python
list(df.columns)
```

```python
df.describe()
```

# Check if the number of nuclei and foci are within normal limits

```python
import warnings
warnings.filterwarnings('ignore')

fig, ax = plt.subplots()
nuclei_count_bxplt = df.loc[(df.Count_Nuclei.notnull()) ,'Count_Nuclei']
h2ax_count_bxplt = df.loc[(df.Count_h2AX.notnull()),'Count_h2AX']
ax.boxplot([nuclei_count_bxplt, h2ax_count_bxplt])
ax.set(xticklabels=['Nuclei', 'H2AX Foci'], title="Numbers of Objects")
plt.show()
```

# Make new features

```python
df['PercentPosNuc'] = (df['Count_RO_PositiveNuclei'] / df['Count_Nuclei']) *100

df['Foci_nuc_ratio'] = df['Count_h2AX'] / df['Count_Nuclei'] 

df[['PercentPosNuc','Foci_nuc_ratio'] ]
df
```

```python
#groups
df_groups = df.groupby('Metadata_Treatment')

df_controls = df_groups.get_group('Control'.upper())
df_PGVG = df_groups.get_group('PGVG')
df_Nic1 = df_groups.get_group('NIC1')
df_Nic2 = df_groups.get_group('NIC2')
df_Nic3 = df_groups.get_group('NIC3')
df_CSE = df_groups.get_group('CSE')

df_controls
```

```python
import warnings
warnings.filterwarnings('ignore')

fig, ax = plt.subplots()
control_nuclei_count_bxplt = df_controls.loc[(df_controls.Count_Nuclei.notnull()) ,'Count_Nuclei']
PGVG_nuclei_count_bxplt = df_PGVG.loc[(df_PGVG.Count_Nuclei.notnull()) ,'Count_Nuclei']
Nic1_nuclei_count_bxplt = df_Nic1.loc[(df_Nic1.Count_Nuclei.notnull()) ,'Count_Nuclei']
Nic2_nuclei_count_bxplt = df_Nic2.loc[(df_Nic2.Count_Nuclei.notnull())  ,'Count_Nuclei']
Nic3_nuclei_count_bxplt = df_Nic3.loc[(df_Nic3.Count_Nuclei.notnull())  ,'Count_Nuclei']
CSE_nuclei_count_bxplt = df_CSE.loc[(df_CSE.Count_Nuclei.notnull())  ,'Count_Nuclei']
ax.boxplot([control_nuclei_count_bxplt, PGVG_nuclei_count_bxplt, Nic1_nuclei_count_bxplt, Nic2_nuclei_count_bxplt, Nic3_nuclei_count_bxplt, CSE_nuclei_count_bxplt])
ax.set(xticklabels = ['Control', 'PGVG', '12.5mg', '25mg', '50mg', 'CSE'], title="Count_Nuclei")
plt.show()
```

# QC

```python
#Control
'''
We are getting the Extract IDs for CONTROL
'''

ExtractQC_control = df[['Metadata_Treatment', 
                        'Metadata_ExtractID', 
                        'PercentPosNuc',
                        'Foci_nuc_ratio',
                        'Mean_h2AX_AreaShape_Area',
                        'Mean_h2AX_AreaShape_MeanRadius',
                        'Mean_h2AX_Intensity_MeanIntensity_MaskedGreen',
                        'Mean_h2AX_Intensity_MedianIntensity_MaskedGreen',
                        'ImageQuality_Correlation_h2AX_10',
                        'ImageQuality_PercentMaximal_h2AX',
                        'ImageQuality_ThresholdRobustBackground_h2AX',
                        'Threshold_FinalThreshold_h2AX']].groupby('Metadata_Treatment').get_group('CONTROL')

extractIds = ExtractQC_control['Metadata_ExtractID'].unique() #will be used as labels
print(extractIds)


rows = []
columns = ['Metadata_ExtractID','Average_PercentPosNuc','Std_Dev_PercentPosNuc', 'Mean_Foci_nuc_ratio','Std_Dev_Foci_nuc_ratio','Sample_Size'] #for new dataframe

for extractId in extractIds:
    print('Looking at extract ID ',extractId)
    mean_percentPosNuc = ExtractQC_control.loc[ ExtractQC_control['Metadata_ExtractID']==extractId]['PercentPosNuc'].mean() #getting the mean for the extract Id
    print(mean_percentPosNuc)
    
    std_dev_percentPosNuc = ExtractQC_control.loc[ ExtractQC_control['Metadata_ExtractID']==extractId]['PercentPosNuc'].std()
    sample_size = len(ExtractQC_control.index)
    
    Mean_Foci_nuc_ratio = ExtractQC_control.loc[ ExtractQC_control['Metadata_ExtractID']==extractId]['Foci_nuc_ratio'].mean() #getting the mean for the extract Id
    print(mean_percentPosNuc)
    
    Std_Dev_Foci_nuc_ratio = ExtractQC_control.loc[ ExtractQC_control['Metadata_ExtractID']==extractId]['Foci_nuc_ratio'].std()
    sample_size = len(ExtractQC_control.index)
    
    
    rows.append([extractId,mean_percentPosNuc,std_dev_percentPosNuc,Mean_Foci_nuc_ratio,Std_Dev_Foci_nuc_ratio,sample_size]) #add to new dataframe


Control_Extracts = pd.DataFrame(rows,columns=columns)
Control_Extracts

#make bar charts
labels = extractIds
x = np.arange(4)
width = 0.25

fig, ax = plt.subplots(5,2, figsize=(15,20))
ax[0,0].bar(x,Control_Extracts['Average_PercentPosNuc'],width,label='Percent Pos Nuc',color='blue',yerr=(Control_Extracts['Std_Dev_PercentPosNuc']))
ax[0,0].set_xticks(x)
ax[0,0].set_xticklabels(labels)
ax[0,0].set_ylabel('Average Percent Positive Nuclei')
ax[0,0].set_xlabel('Extract ID')
ax[0,0].legend()
ax[0,0].set_title('Control ')

ax[0,1].bar(x,Control_Extracts['Mean_Foci_nuc_ratio'],width,label='Foci Nuc Ratio',color='green',yerr=(Control_Extracts['Std_Dev_Foci_nuc_ratio']))
ax[0,1].set_xticks(x)
ax[0,1].set_xticklabels(labels)
ax[0,1].set_ylabel('Mean Foci Nuclei ratio')
ax[0,1].set_xlabel('Extract ID')
ax[0,1].legend()

ax[1,0].hist(ExtractQC_control['ImageQuality_Correlation_h2AX_10'],50)
ax[1,0].set_xlabel('Correlation')

ax[1,1].hist(ExtractQC_control['ImageQuality_PercentMaximal_h2AX'],50)
ax[1,1].set_xlabel('%maximal')

ax[2,0].hist(ExtractQC_control['ImageQuality_ThresholdRobustBackground_h2AX'],50)
ax[2,0].set_xlabel('QC TH')

ax[2,1].hist(ExtractQC_control['Mean_h2AX_AreaShape_MeanRadius'],50)
ax[2,1].set_xlabel('Mean_h2AX_AreaShape_MeanRadius')

ax[3,0].hist(ExtractQC_control['Mean_h2AX_AreaShape_Area'],50)
ax[3,0].set_xlabel('Mean_h2AX_AreaShape_Area')

ax[3,1].hist(ExtractQC_control['Mean_h2AX_Intensity_MeanIntensity_MaskedGreen'],50)
ax[3,1].set_xlabel('Mean_h2AX_Intensity_MeanIntensity_MaskedGreen')

ax[4,0].hist(ExtractQC_control['Mean_h2AX_Intensity_MedianIntensity_MaskedGreen'],50)
ax[4,0].set_xlabel('Mean_h2AX_Intensity_MedianIntensity_MaskedGreen')


ax[4,1].scatter(ExtractQC_control['Mean_h2AX_AreaShape_Area'],ExtractQC_control['Mean_h2AX_Intensity_MeanIntensity_MaskedGreen'])
ax[4,1].set_ylabel('MeanIntensity')
ax[4,1].set_xlabel('Mean_h2AX_AreaShape_Area')
```
